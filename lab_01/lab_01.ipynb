## Styles and paradigms of computer programming

The general idea of computer programming can be realized in many different ways and perhaps more importantly at many different levels of abstraction. Different problems and application domains may require different approaches, so there is no universally best one-size-fit-all language/approach. This is why we have such a vast array of computer languages and frameworks to choose from. Moreover, although in principle any proper programming language can be used to realize any kind of programming paradigm/framework, in practice some of them are much better suited for some problems and not for others.

It is then important to be aware of the main ways in which the practice of computer programming can be conceived of. In other words it is important to know differences between different **programming paradigms** and to know when each of them should be used.

### Program as a recipie

<img src="recipes.png" width="300px" />

- A program defines a sequence of operations to be carried out by a machine.
- Actions are executed **line-by-line** in a fixed linear fashion.
- Each instruction corresponds to a particular operation.

**Application:** low-level programming (e.g. microcontrollers, OS programming).

**Advantages:** control over low-level details of how every step of a procedure is implemented.

**Drawbacks:** very verbose code, caring about all low-level details makes writing complex applications very hard and error-prone.

**Styles:** imperative programming (procedural, structured).

### Program as a description

<img src="plan.jpg" width="300px" />

- A program desribes the desired outcome not the exact way in which it has to be produced.
- Ordering of lines of code does not have to map 1-to-1 to the order of execution.
- A program is a model: its structure reflect particular properties of the modeled phenomena.

**Applications:** process/object modeling.

**Advantages:** abstract, high-level nature makes description of complex processes easier.

**Drawbacks:** little control over low-level details of how particular operations are implemented.

**Styles:** declarative programming (e.g. database query languages), object-oriented programming.

### Program as a set of specialzed modules 

<img src="modules.jpg" width="300px" />

- A program is decomposed into set of subprograms, which are as decoupled as possible (i.e. changes in one of them do not affect the rest).
- Each part (module) can be considered an independent entity, so they can be freely rearranged and composed to make larger structures.
- Communication between modules is possible through highly formalized interfaces.
- Modules control the flow of actions.

**Application:** modeling of complex systems with multiple interacting parts, computer simulations, large and complex software systems.

**Advantages:** modularity makes it easy to maintain even very large and complex projects as each subpart can be modified without affecting other parts.

**Drawbacks:** modules have to be able to communicate and coordinate what adds an additional layer of complexity to the entire system as well as may make it less performant in terms of execution times.

**Styles:** object-oriented programming, aspect programming, functional programming.

### Program as a data processing pipeline

<img src="pipes.jpg" width="300px" />

- A program controls the flow of data and transform the input into a desired output data.
- The strucutre of a program reflects input-output relationships between subsequent transformations.
- Properties of data controls the flow of a program.
- Applications have no internal state. Everything that happens follows from defintitions of functions and properties of data.

**Applications:** scientific computing / statistics, machine learning, parallel and distributed computing.

**Advantages:**  no internal state makes it easier to debug applications and enforces decomposition of complex processes into functional parts.

**Drawbacks:** lack of internal state makes it harder to express particular types of computations.

**Styles:** data-driven programming, array processing, functional programming.


## Reality

Having said all of the above, it is important to remember that in real world applications quite often multiple paradigms are mixed together and that too much of purity of style often simply leads to bad code. Real-world applications are usually complex enough to be composed of several subproblems for which different approaches may be optimal. Moreover, in some sense all paradigms are connected as ultimately all code reduced to linear set of operations expressed in machine code (the only language that computers really speak).

In the context of typical scientific / statistical computing (the application domain that we will focus on in this class) programs usually are structured using a mix of procedural/functional/object-oriented programming patterns with the prevalence of procedural/functional. This is due to the simple fact that most (although not all) scientific problems are largely data-driven and as such can be naturally decomposed into a sequence of computations performed on data that can be carried out one by one.

## Basics of OOP

The very idea behind the object-oriented programming is to write programs in such a way so they model the external world and objects are meant to reflect parts of this world, especially those that play important roles in the problem at hand. As a result we can usually think about objects as characters. Every such character has some properties or attributes which describe it and some skills or abilities that define the actions that he or she can perform.

So in the OOP it is the same. Objects are in fact containers that store some data (**properties**) and procedures for operating on this data and optionally some additional input. These procedures defined on objects are called **methods**. We introduce these notions because we will not be able to avoid them.

In other words properties and methods mirror the more fundamental notions of a variable and a function, because in some sense they are exactly variables and functions but defined on an object, so they are decoupled from the external environment (each object defines a separate namespace). That means that we can have only one global function of a given name, but multiple methods with the same name as long as they are defined on different objects.

Objects are also directly related to the fundamental notion of a class. A class is a generic type of an object from which all objects of a given type are created. In some sense a class is a template or a blueprint for creating obejcts. An object that belongs to some class is often called an **instance** of that class.

````python
class Character:
    """This is a template for creating character objects.
    
    It has one attribute `name` which defines the name of the character.
    It also defines one method `greet` that prints a nice welcome message
    from the character that includes his or her name.
    
    Besides the `greet` method there is also this weird `__init__` method.
    This is so-called initialization method which is used to setup an instance
    of the class, that is, assign values to properties that have to be defined
    from the get go.
    
    Notice that all methods defined for a class of objects have to take
    a special argument `self` at the first position. This `self` object
    is a reference to the object itself, so it can freely use its own
    attributes and other methods within any given method.
    """
    
    def __init__(self, name):
        """Initialization method.
        
        Takes one argument `name` and assigns its value to the property
        `name` defined on `self`.
        """
        self.name = name
    
    def greet(self):
        """Prints a welcome message and takes the value of name from `self`."""
        print("Greetings stranger! I am {}.".format(self.name))

def greet(name):
    """This does not mess up with the `greet` method of the character class
    as it lives in a different namespace.
    """
    print("Greetings stranger! I am {}.".format(name))
````

````python
minsc = Character("Minsc")
minsc.greet()
greet("Minsc")
````

We will not go into more details of OOP. But we need to be aware of the fundamental notions of:
